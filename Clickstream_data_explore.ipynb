````python
# preprocessing clickstream data
df_clickstream=df_clickstream_tmp.iloc[:, [1,3,4,5,6,7]]
df_clickstream.head(5)
# df_clickstream['API_key'].nunique()
````

````python
#remove www.
df_clickstream['domain']=df_clickstream['domain'].str.replace("www.", "")
# remove safebrowsing-cache. is this useful/appropriate?
df_clickstream['domain']=df_clickstream['domain'].str.replace("safebrowsing-cache.", "")
# # # remove nl-nl. is this useful/appropriate?
df_clickstream['domain']=df_clickstream['domain'].str.replace("nl-nl.", "")

df_clickstream['domain'].sample(5)
# df_clickstream['API_key'].nunique()
````

````python
# loading white list / Categories lis
df_white_list = pd.read_csv("nadia_whitelist.csv", encoding = "ISO-8859-1") 
df_white_list.sample(5)
````

````python
# to a data frame df_clickstream
df_rob= pd.read_csv("rob_categor.csv") 
# df_rob['API_key'].nunique()
````

````python
df_rob_category = df_rob[['API_key', 'cp.code', 'page_url']].rename(columns={'cp.code':'code'})
df_rob_category
# df_rob_category['API_key'].nunique()
````

````python
df_rob_category['code'].str.count("N1").sum()
````

````python
df_rob_category['code'].str.count("N2").sum()
````

````python
df_rob_category['code'].size
# df_rob_category['API_key'].nunique()
````

````python
df_browsing_tmp=df_rob_category.groupby(['API_key','code']).size().reset_index(name='Total')
df_browsing_tmp.head(5)
````

````python
df_total_browsing = df_browsing_tmp.groupby(['API_key']).sum().reset_index()
df_total_browsing.head(10)
````

````python
df_N1 = df_browsing_tmp[df_browsing_tmp.code == 'N1']
df_N1 = df_N1[['API_key','Total']].rename(columns={'Total':'N1'})

df_N2 = df_browsing_tmp[df_browsing_tmp.code == 'N2']
df_N2 = df_N2[['API_key','Total']].rename(columns={'Total':'N2'})


df_X = df_browsing_tmp[df_browsing_tmp.code == 'X']
df_X = df_X[['API_key','Total']].rename(columns={'Total':'X'})

df_G = df_browsing_tmp[df_browsing_tmp.code == 'G']
df_G = df_G[['API_key','Total']].rename(columns={'Total':'G'})

df_S = df_browsing_tmp[df_browsing_tmp.code == 'S']
df_S = df_S[['API_key','Total']].rename(columns={'Total':'S'})
# df_S.head(20)
````

````python

df_browsing = pd.merge(left=df_total_browsing, right=df_N1, how='left', left_on='API_key', right_on='API_key')

df_browsing = pd.merge(left=df_browsing, right=df_N2, how='left', left_on='API_key', right_on='API_key')

df_browsing = pd.merge(left=df_browsing, right=df_X, how='left', left_on='API_key', right_on='API_key')

df_browsing = pd.merge(left=df_browsing, right=df_G, how='left', left_on='API_key', right_on='API_key')

df_browsing = pd.merge(left=df_browsing, right=df_S, how='left', left_on='API_key', right_on='API_key')

df_browsing.head(5)

````

````python
df_browsing=df_browsing.fillna(0)
df_browsing.head(10)
````

````python
df_browsing['N'] = df_browsing['N1'] + df_browsing['N2']
df_browsing['N1%'] = 100 * df_browsing['N1'] / df_browsing['N']
df_browsing['N2%'] = 100 * df_browsing['N2'] / df_browsing['N']
df_browsing.head(5)
````

````python
df_browsing = df_browsing [['API_key', 'N1', 'N1%', 'N2', 'N2%', 'N', 'X', 'G', 'S', 'Total']]
df_browsing.head(5)
````

````python
df_browsing.to_csv("browsing_volume_and_percentage_by_user_by_category.csv")
````
